```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.plot([1,2,3],[110,130,130])
plt.show()
```

```python
plt.plot(['Seoul','Tokyo','Paris'],[31,32,28])
plt.xlabel('City')
plt.ylabel('Temperature (C)')
plt.title('Temperature')
plt.show()
```

```python
plt.plot([1,2,3],[3,4,5],'r-.')
plt.plot([1,2,3],[1,7,6],'b--')
plt.xlabel('Sequence')
plt.ylabel('Time(secs)')
plt.title('Experiment Result')
plt.legend(['Mouse','Cat'])
plt.show()
```

```python
y=[5,4,3,1,8,11,3,9,3]
x=range(len(y))
plt.bar(x,y,width=0.7,color='blue')
plt.show()
```

```python
x=np.arange(0,5,0.01)
plt.plot(x,x**2,'r--')
plt.plot(x,2**x,'b')
plt.axis([0,6,0,50])
plt.show()
```

```python
x=np.arange(0,5,0.01)
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.axis([0,5,0,25])

plt.plot(x,2*x)
plt.subplot(3,2,3)
plt.axis([0,5,0,25])

plt.plot(x,x**2)
plt.subplot(3,2,4)
plt.axis([0,5,0,25])

plt.plot(x,x**x)
plt.subplot(3,2,5)
plt.axis([0,5,0,25])

plt.plot(x,np.log(x))
plt.subplot(3,2,6)
plt.axis([0,5,0,25])

plt.plot(x,np.exp(x))
plt.axis([0,5,0,25])
plt.show()
```

```python
x=np.arange(0,5,0.01)

plt.title('two graph')
plt.plot(x,x,'g',label='y=x')
plt.plot(x,2*(x+1),'r',label='y=2*(x+1)')

plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()

plt.grid(True)
plt.show()
```

```python
x1=[1,2,11,20,6,9]
y1=[3,7,10,20,16,30]

x2=[5,7,1,6,9,2]
y2=[3,8,10,23,4,32]

plt.title('two scatter')
plt.scatter(x1,y1,marker='s',c='r')
#plt.plot(x1,y1,marker='*',c='b')
plt.scatter(x2,y2,marker='*',c='b')

plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x1=[1,2,11,20,6,9]
y1=[3,7,10,20,16,30]

bboxType=dict(boxstyle="round4",alpha=0.8,fc='w')
arrow=dict(arrowstyle='-|>',connectionstyle='arc3,rad=0.2',fc='w')

plt.title('annotation')
plt.scatter(x1,y1,marker='*',c='r',s=40)

plt.annotate('here!',xy=(11,10),xytext=(15,15),size=20,
             ha='center',va='center',
             #horizontal arrow
             arrowprops=arrow, bbox=bboxType
             )

plt.grid(True)
plt.show()
```
