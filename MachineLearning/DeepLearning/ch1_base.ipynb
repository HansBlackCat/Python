```python
# Array from 0 to 6 then 0.1
x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x") # x-axis name
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
