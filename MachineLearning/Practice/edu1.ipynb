```python
# 외부의 라이브러리를 쓰겠다는 뜻
# 라이브러리는 대충 쓸만한 것 모음집 같은것
# import {라이브러리 이름} as {약자} 같이 씀
import numpy as np # 넘파이 - 선형대수나 기본적인 수학
import matplotlib.pyplot as plt # 파이플롯 - 그래프 그려주는 라이브러리
```

```python
# np 로 시작하는 건 우리가 위에서 약자로 np 를 쓰겠다고 해서
# arange 는 A 에서 B 까지 C 간격으로 넘파이 리스트를 만들겠다는 뜻
xvals = np.arange(0,10,0.1) # 0, 0.1, 0.2, ... 9.7, 9.8, 9.9 까지 만들어줌
# 마지막 숫자 10은 안 만들어 준다 조심
yvals = np.exp(xvals) # e^x

# 그림 그림
plt.plot(xvals,yvals)
# 보여줘
plt.show()
```

```python
# 파란색 점으로 그려줘
# '.b' 에서 b 가 blue, . 이 점선이라는 뜻 
plt.plot(xvals,yvals,'.b')
plt.show()
```

```python
# 여러개 그리기
fig, ax = plt.subplots()

y2 = np.exp(-(xvals-10))

ax.plot(xvals, yvals, '-k') # 검은 선
ax.plot(xvals, 1500*xvals, '.r') #붉은 점
ax.plot(xvals, y2, '*c') # cyan 색 별
```

```python
# 여러개 그리기 2

# 랜덤 숫자 0~1 까지 10개 뽑기
x = np.random.rand(10)
y = np.random.rand(10)
# z = sqrt(x^2 + y^2) - 원점으로부터 거리
# ** 곱하기 2번한게 제곱
z = np.sqrt(x**2 + y**2)

# 매트랩 처럼 여러개 그리기

# 큰 제목 쓰기 어렵다

# 321 -> 3x2 모양으로 그리는 그래프중 1번 째
plt.subplot(321)
# 산포도 그리기, s = size, c = color
plt.scatter(x, y, s=80, c=z, marker=">")
plt.ylabel("hello")


plt.subplot(322)
#                            대충 오각형으로 그리라는 뜻
plt.scatter(x, y, s=80, c=z, marker=(5, 0))


# 그림 그리기 저 4점 이으면 3번째 그림 됨
verts = np.array([[-1, -1], [1, -1], [1, 1], [-1, -1]])
plt.subplot(323)
#                            그림그린 대로 그리기
plt.scatter(x, y, s=80, c=z, marker=verts)


tmp = np.arange(0,1,0.1)
plt.subplot(324)
#                  5각 v2
plt.plot(tmp, tmp, marker=(5, 1))


plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')


plt.subplot(326)
#                            5각 v3
plt.scatter(x, y, s=80, c=z, marker=(5, 2))

plt.show()
```

```python
def f(t):
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return s1 * e1

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 2.0, 0.01)

# 아까 위에도 있던건데
# fig 는 전체 plot
# ax 는 subplot 이라 생각하면 편함
#                             이건 대충 글자처럼 취급 같은거, 안겹치게 함
fig, axs = plt.subplots(2, 1, constrained_layout=True)

axs[0].plot(t1, f(t1), 'o', t2, f(t2), '-')
axs[0].set_title('subplot 1')
axs[0].set_xlabel('distance (m)')
axs[0].set_ylabel('Damped oscillation')
fig.suptitle('This is a somewhat long figure title', fontsize=16)

axs[1].plot(t3, np.cos(2*np.pi*t3), '--')
axs[1].set_xlabel('time (s)')
axs[1].set_title('subplot 2')
axs[1].set_ylabel('Undamped')

plt.show()
```
